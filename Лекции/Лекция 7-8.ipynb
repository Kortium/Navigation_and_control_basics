## Общие понятия и классификация

#### Основные категории беспилотных систем
    
- **Летательные беспилотные системы**
    - Примеры: квадрокоптеры, фиксированные крылья, гибридные платформы.
    - Ключевые особенности:
        - Высокий уровень маневренности и требования к параметрам ориентации.
        - Более сложная система управления из-за требований к тяге и аэродинамике.
    - Применение: съемка, картографирование, доставка.

- **Наземные беспилотные системы**
    - Примеры: автономные автомобили, роботы доставщики, исследовательские платформы для труднодоступных мест.
    - Ключевые особенности:
        - Устойчивость и маневренность на различных типах поверхности.
        - Низкий уровень энергии, потребляемой на поддержание положения.
    - Применение: логистика, сельское хозяйство, спасательные операции.

- **Необитаемые подводные аппараты**
    - Ключевые особенности:
        - Высокий требования к автономности подводных систем.
    - Применение: картографирование, спасательные операции.

#### Основные задачи управления
- **Стабилизация положения**
    - Задача: удержание объекта в определенном положении и ориентации в пространстве.
    - Применение: дрон в состоянии зависания, автомобиль следующий с заданными скоростями.
    - Методы стабилизации: использование данных от сенсоров (гироскопов, акселерометров) и контроллеров, таких как PID.
- **Контроль траектории движения**
    - Задача: управление движением вдоль заданного пути или к определенной точке.
    - Применение: наземный робот, следящий за траекторией на складе; дрон, перемещающийся по маршруту.
    - Алгоритмы: Pure Pursuit, DWA (Dynamic Window Approach), PID-контроллер.
- **Управление состоянием системы**
    - Задача: мониторинг и контроль определенных параметров системы, например, заряд батареи, датчики температуры и приводы.
    - Применение: управление потреблением энергией в дронах для увеличения времени полета, контроль температуры двигателей.
    - Методы: адаптивное управление, использование систем обратной связи и диагностики.

#### Актуаторы
Актуатор — это устройство, которое преобразует сигналы управления в механическое движение.

### Основные функции актуаторов в беспилотных системах:
1. **Приведение системы в движение**: актуаторы управляют движением системы (вращение колес, винтов, лопастей и т.д.).
2. **Контроль ориентации и стабилизация**: актуаторы помогают удерживать объект в заданном положении и ориентации.
3. **Маневрирование**: выполняют различные маневры и повороты, управляют скоростью и направлением.

Рассмотрим актуаторы для различных типов беспилотных систем: летательных аппаратов, наземных беспилотных систем и подводных аппаратов.

### 1. Актуаторы для летательных аппаратов (дронов и других беспилотных воздушных систем)

Для летательных аппаратов основными актуаторами являются:
- **Пропеллеры (винты)**, приводимые в движение электродвигателями. Они создают тягу, поднимая и удерживая дрон в воздухе, а также управляют его движением.
- **Электродвигатели** — используются для вращения пропеллеров. Изменение скорости вращения двигателя приводит к изменению тяги, что позволяет контролировать высоту и направление полета.
- **Рули направления (в аппаратах с крыльями)** — управляющие поверхности (например, рули высоты и направления) позволяют управлять креном, тангажом и рысканием.

**Принципы работы и управления**:
- Для удержания дрона на месте требуется одинаковая тяга от всех пропеллеров. Для выполнения маневров тяга одного или нескольких пропеллеров меняется, создавая вращение и наклон в нужном направлении.
- В квадрокоптерах для стабилизации используются два вращающихся по часовой стрелке и два — против часовой. Это позволяет сбалансировать крутящий момент и стабилизировать положение.

### 2. Актуаторы для наземных колесных беспилотных систем

В колесных системах актуаторы предназначены для управления движением и скоростью. Основными актуаторами для таких систем являются:
- **Колеса, приводимые в движение электродвигателями или мотор-колесами**: основные элементы для передвижения по поверхности.
- **Сервоприводы для рулевого управления** — используются для поворота колес и маневрирования.

**Принципы работы и управления**:
- Управление скоростью и направлением движения осуществляется изменением скорости и направления вращения колес.
- Для маневров электродвигатели могут регулировать обороты индивидуально для каждого колеса (как в дифференциальных системах управления).
- Сервоприводы обеспечивают поворот колес для изменения траектории.

### 3. Актуаторы для подводных аппаратов

Основные актуаторы для подводных аппаратов:
- **Гребные винты (Thrusters)** — электродвигатели, которые создают тягу для передвижения в воде.
- **Сервоприводы для управления погружением и перемещением**: выполняют повороты и погружение в воду.
- **Системы балластных цистерн**: позволяют регулировать плавучесть и глубину погружения.

**Принципы работы и управления**:
- Винты работают аналогично пропеллерам дронов, создавая тягу для перемещения вперед, назад или в стороны.
- Для погружения и всплытия могут применяться балластные системы или вертикальные винты.
- Маневры, такие как повороты и наклоны, выполняются с помощью изменения тяги в различных направлениях.

### Сравнение актуаторов для различных типов беспилотных систем

| Тип системы              | Основные актуаторы        | Функции                                               |
|--------------------------|---------------------------|-------------------------------------------------------|
| Летательные аппараты     | Пропеллеры, электродвигатели, рули | Создание тяги, стабилизация и маневрирование          |
| Наземные колесные системы | Колеса, электродвигатели, сервоприводы | Перемещение по поверхности, управление поворотом     |
| Подводные аппараты       | Винты, балластные системы, сервоприводы | Движение в воде, погружение, всплытие, маневрирование |

Каждый из этих типов актуаторов имеет особенности, связанные с физическими требованиями среды, в которой действует беспилотная система, и типом задач, которые нужно выполнить.

### Базовые контроллеры

**PID-контроллер** (пропорционально-интегрально-дифференциальный контроллер) — это один из наиболее распространенных типов регуляторов в системах автоматического управления. Он корректирует управляющее воздействие на объект на основе трех составляющих:

1. **Пропорциональная (P) составляющая**:
   - Реагирует на текущую ошибку $ e(t) $ между желаемым значением и текущим измеренным значением.
   - Управляющее воздействие:
     $
     U_P(t) = K_P \cdot e(t)
     $
   - Где $ K_P $ — коэффициент пропорциональности.

2. **Интегральная (I) составляющая**:
   - Учитывает накопленную ошибку за время.
   - Управляющее воздействие:
     $
     U_I(t) = K_I \int_{0}^{t} e(\tau) d\tau
     $
   - Где $ K_I $ — интегральный коэффициент.
   - Помогает устранить постоянную систематическую ошибку (смещение).

3. **Дифференциальная (D) составляющая**:
   - Реагирует на скорость изменения ошибки.
   - Управляющее воздействие:
     $
     U_D(t) = K_D \cdot \frac{de(t)}{dt}
     $
   - Где $ K_D $ — дифференциальный коэффициент.
   - Предотвращает перерегулирование и улучшает быстродействие системы.

**Общее управляющее воздействие**:

$
U(t) = U_P(t) + U_I(t) + U_D(t) = K_P e(t) + K_I \int_{0}^{t} e(\tau) d\tau + K_D \frac{de(t)}{dt}
$

##### Настройка PID-контроллера

Настройка коэффициентов $ K_P $, $ K_I $, $ K_D $ является критически важной для эффективной работы контроллера. Основные методы настройки:

- **Метод Циглера-Никольса**: Эмпирический метод, использующий реакцию системы на ступенчатое воздействие.
- **Автоматизированные методы**: Использование программных средств для подбора оптимальных параметров.
- **Эволюционные алгоритмы**: Применение генетических алгоритмов и других методов оптимизации.

##### Варианты PID-контроллеров

- **P-контроллер**: Использует только пропорциональную составляющую. Прост в реализации, но может иметь постоянную ошибку.
- **PI-контроллер**: Комбинация пропорциональной и интегральной составляющих. Устраняет постоянную ошибку, но может быть медленнее.
- **PD-контроллер**: Использует пропорциональную и дифференциальную составляющие. Быстрее реагирует на изменения, но может иметь постоянную ошибку.

#### Пример

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import odeint

class PIDController:
    def __init__(self, Kp=0, Ki=0, Kd=0, setpoint=0):
        self.Kp = Kp  # Пропорциональный коэффициент
        self.Ki = Ki  # Интегральный коэффициент
        self.Kd = Kd  # Дифференциальный коэффициент
        self.setpoint = setpoint  # Заданное значение
        self.integral = 0  # Интегральная сумма ошибок
        self.prev_error = 0  # Предыдущая ошибка

    def update(self, feedback_value, dt):
        error = self.setpoint - feedback_value
        self.integral += error * dt
        derivative = (error - self.prev_error) / dt if dt > 0 else 0
        output = self.Kp * error + self.Ki * self.integral + self.Kd * derivative
        self.prev_error = error
        return output


def simulate_system(Kp, Ki, Kd, title):
    # Начальные условия
    y = [0.0, 0.0]  # y[0]: положение, y[1]: скорость
    pid = PIDController(Kp=Kp, Ki=Ki, Kd=Kd, setpoint=1.0)
    t_max = 10.0  # Время симуляции
    dt = 0.01  # Шаг интегрирования
    n_steps = int(t_max / dt)
    # Массивы для хранения результатов
    t_array = np.linspace(0, t_max, n_steps)
    y_array = np.zeros((n_steps, 2))
    u_array = np.zeros(n_steps)
    # Симуляция
    for i in range(n_steps):
        t = i * dt
        # Обновление контроллера и получение управляющего воздействия
        u = pid.update(y[0], dt)
        # Запись результатов
        y_array[i, :] = y
        u_array[i] = u
        # Вычисление производных
        dy1_dt = y[1]
        dy2_dt = u - 2 * y[1] - y[0]
        # Обновление состояния методом Эйлера
        y[0] += dy1_dt * dt
        y[1] += dy2_dt * dt
    # Построение графиков
    plt.figure(figsize=(10, 5))
    plt.plot(t_array, y_array[:, 0], label='Положение')
    plt.plot(t_array, [pid.setpoint]*len(t_array), 'r--', label='Заданное значение')
    plt.title(title)
    plt.xlabel('Время, с')
    plt.ylabel('Положение')
    plt.legend()
    plt.grid()
    plt.show()
```

```python
simulate_system(Kp=5.0, Ki=0.0, Kd=0.0, title='Только P-контроллер (Kp=5.0)')
```

```python
simulate_system(Kp=5.0, Ki=1.0, Kd=0.0, title='PI-контроллер (Kp=5.0, Ki=1.0)')
```

```python
simulate_system(Kp=25.0, Ki=0.0, Kd=0.0, title='PID-контроллер (Kp=5.0, Ki=1.0, Kd=1.0)')
```

### Траекторное управление
Задача траекторного управления заключается в формировании управляющих воздействий для аппараты, чтобы он следовала заданной траектории или пути во времени с определенными требованиями к точности и динамике.

#### Составные части задачи траекторного управления

1. **Модель объекта управления**:
   - **Кинематическая модель**: Описывает движение объекта без учета сил и моментов.
   - **Динамическая модель**: Учитывает силы, моменты, инерцию и другие динамические эффекты. Необходима для более точного управления.

2. **Заданная траектория**:
   - **Позиция во времени**: Набор точек $ \mathbf{r}(t) = [x(t), y(t), z(t)] $, которые объект должен пройти.
   - **Скорость и ускорение**: Требования к компонентам скорости $ \mathbf{v}(t) $ и ускорения $ \mathbf{a}(t) $ на траектории для обеспечения плавности и динамики движения.

3. **Система управления**

4. **Обратная связь**:
   - **Сенсоры и измерения**: Средства получения информации о текущем состоянии объекта (позиция, скорость, ориентация).

5. **Критерии качества управления**:
   - **Точность следования**: Максимальное допустимое отклонение от заданной траектории.
   - **Стабильность**: Способность системы возвращаться к траектории после возмущений.
   - **Оптимальность**: Минимизация затрат энергии, времени или других ресурсов при следовании траектории.

6. **Ограничения системы**:
   - **Физические ограничения**: Максимальные скорости, ускорения, силы и моменты, доступные актуаторам.
   - **Окружающая среда**: Препятствия, условия сцепления, ветровые нагрузки и другие внешние факторы.

7. **Планировщик траекторий**:
   - **Генерация траекторий**: Создание траекторий, учитывающих динамику объекта и ограничения.
   - **Перепланирование**: Способность изменять траекторию в реальном времени при возникновении препятствий или изменений в среде.

### Алгоритмы траекторного управления

#### Алгоритм DWA (Dynamic Window Approach)

**Dynamic Window Approach (DWA)** — это метод локального планирования движения для мобильных роботов, который позволяет избегать препятствий и двигаться к цели, учитывая динамические ограничения робота.

##### Принцип работы

1. **Определение динамического окна**:
   - Динамическое окно представляет собой набор допустимых скоростей $ (v, \omega) $, которые робот может достичь за короткий промежуток времени с учетом его максимальных ускорений и скоростей.

2. **Генерация траекторий**:
   - Для каждого сочетания скоростей внутри динамического окна прогнозируется траектория движения на короткий период времени.

3. **Оценка траекторий**:
   - Каждая траектория оценивается по критериям:
     - **Безопасность**: Расстояние до ближайших препятствий.
     - **Целеустремленность**: Насколько траектория приближает робота к цели.
     - **Скорость**: Предпочтение более высоких скоростей для эффективности.

4. **Выбор оптимальной траектории**:
   - Выбирается траектория с наивысшей общей оценкой, и соответствующие скорости применяются к роботу.

##### Преимущества и недостатки

- **Преимущества**:
  - Реактивность: Быстрая адаптация к изменениям в окружающей среде.
  - Учет динамики: Соответствует физическим возможностям робота.
- **Недостатки**:
  - Локальность: Может застрять в локальных минимумах.
  - Требует вычислительных ресурсов для оценки множества траекторий.

```python
# Пример базового управления

from Roko2 import Roko2

# Инициализация робота с начальными условиями
x0 = 0  # начальная координата x
y0 = 0  # начальная координата y
heading0 = 0  # начальный угол поворота (в радианах)
velocity0 = 0  # начальная линейная скорость
angular_rate0 = 0  # начальная угловая скорость

robot = Roko2(x0, y0, heading0, velocity0, angular_rate0)

# Установка желаемых параметров движения
desired_velocity = 1.0  # м/с
desired_angular_rate = 0.1  # рад/с

# Количество шагов симуляции
num_steps = 100

# Цикл симуляции
for _ in range(num_steps):
    # Установка желаемых параметров движения
    robot.set_motion(desired_velocity, desired_angular_rate)
    # Обновление состояния робота
    robot.update()

# Построение результатов
robot.plot_results()
```

### Принципы управления

Управление роботом осуществляется путем приложения моментов $ M_l $ и $ M_r $ к левому и правому колесам соответственно. Цель состоит в том, чтобы робот следовал заданным значениям линейной скорости $ v $ и угловой скорости $ \omega $.

### Законы управления

Для достижения желаемых скоростей используются пропорционально-дифференциальные (ПД) регуляторы как для линейной, так и для угловой скорости.

### Управление линейной скоростью

1. **Ошибка по линейной скорости**:

   $
   e_v = v_{\text{desired}} - v_{\text{measured}}
   $

2. **Изменение ошибки**:

   $
   \Delta e_v = e_v - e_{v_{\text{prev}}}
   $

3. **Управляющее воздействие**:

   $
   u_v = K_p^v \cdot e_v + K_d^v \cdot \Delta e_v
   $

   Где:
   - $ K_p^v $ — коэффициент пропорциональной составляющей для линейной скорости.
   - $ K_d^v $ — коэффициент дифференциальной составляющей для линейной скорости.

### Контроль угловой скорости

1. **Ошибка по угловой скорости**:

   $
   e_\omega = \omega_{\text{desired}} - \omega_{\text{measured}}
   $

2. **Изменение ошибки**:

   $
   \Delta e_\omega = e_\omega - e_{\omega_{\text{prev}}}
   $

3. **Управляющее воздействие**:

   $
   u_\omega = K_p^\omega \cdot e_\omega + K_d^\omega \cdot \Delta e_\omega
   $

   Где:
   - $ K_p^\omega $ — коэффициент пропорциональной составляющей для угловой скорости.
   - $ K_d^\omega $ — коэффициент дифференциальной составляющей для угловой скорости.

### Формирование управляющих моментов

Полученные управляющие воздействия используются для вычисления моментов на колесах:

$
M_l = \frac{u_v - u_\omega}{2}
$

$
M_r = \frac{u_v + u_\omega}{2}
$

Это позволяет независимо управлять как линейной скоростью, так и угловой скоростью робота.

### Добавим к базовом управлению траекторное

```python
class DWAPlanner:
    def __init__(self, config):
        self.config = config

    def plan(self, robot_state, goal):
        # 1. Генерируем набор допустимых скоростей
        Vs = self.calc_dynamic_window(robot_state)
        # 2. Оценка траекторий
        u, trajectory = self.evaluate_trajectories(robot_state, Vs, goal)
        return u

    def calc_dynamic_window(self, robot_state):
        # Динамическое окно на основе ограничений робота
        config = self.config
        # Текущие скорости
        v = robot_state.velocity
        omega = robot_state.angular_rate
        # Максимальные и минимальные скорости с учетом ускорений
        Vs = [
            max(v - config['max_accel'] * config['dt'], config['min_speed']),
            min(v + config['max_accel'] * config['dt'], config['max_speed']),
            max(omega - config['max_delta_yaw_rate'] * config['dt'], -config['max_yaw_rate']),
            min(omega + config['max_delta_yaw_rate'] * config['dt'], config['max_yaw_rate'])
        ]
        return Vs

    def evaluate_trajectories(self, robot_state, Vs, goal):
        config = self.config
        best_score = float('-inf')
        best_u = [0.0, 0.0]
        best_trajectory = []
        # Перебираем возможные скорости
        for v in np.arange(Vs[0], Vs[1], config['v_reso']):
            for omega in np.arange(Vs[2], Vs[3], config['omega_reso']):
                trajectory = self.predict_trajectory(robot_state, v, omega)
                # Оценка траектории
                to_goal_cost = config['to_goal_cost_gain'] * self.calc_to_goal_cost(trajectory, goal)
                speed_cost = config['speed_cost_gain'] * (config['max_speed'] - v)
                # Общая оценка
                score = to_goal_cost + speed_cost
                if score > best_score:
                    best_score = score
                    best_u = [v, omega]
                    best_trajectory = trajectory
        return best_u, best_trajectory

    def predict_trajectory(self, robot_state, v, omega):
        config = self.config
        x = robot_state.x
        y = robot_state.y
        heading = robot_state.heading
        trajectory = []
        time = 0
        while time <= config['predict_time']:
            x += v * math.cos(heading) * config['dt']
            y += v * math.sin(heading) * config['dt']
            heading += omega * config['dt']
            time += config['dt']
            trajectory.append([x, y, heading, v, omega])
        return trajectory

    def calc_to_goal_cost(self, trajectory, goal):
        # Расстояние от конца траектории до цели
        dx = goal[0] - trajectory[-1][0]
        dy = goal[1] - trajectory[-1][1]
        cost = math.hypot(dx, dy)
        return -cost  # Отрицательное значение, так как мы максимизируем оценку

config = {
    'max_speed': 1.0,         # м/с
    'min_speed': -0.5,        # м/с
    'max_yaw_rate': 40.0 * math.pi / 180.0,  # рад/с
    'max_accel': 1.2,         # м/с^2
    'max_delta_yaw_rate': 180.0 * math.pi / 180.0,  # рад/с^2
    'v_reso': 0.02,           # Разрешение по скорости
    'omega_reso': 1.0 * math.pi / 180.0,  # Разрешение по угловой скорости
    'dt': 0.1,                # Временной шаг для управления
    'predict_time': 2.0,      # Время предсказания траектории
    'to_goal_cost_gain': 1.0, # Коэффициент оценки близости к цели
    'speed_cost_gain': 1.0,   # Коэффициент оценки скорости
}
```

```python
waypoints = [
    [0, 0],
    [5, 0],
    [5, 5],
    [0, 5],
    [0, 0]
]
```

```python
# Инициализация робота с начальными условиями
x0 = 0  # начальная координата x
y0 = 0  # начальная координата y
heading0 = 0  # начальный угол поворота (в радианах)
velocity0 = 0  # начальная линейная скорость
angular_rate0 = 0  # начальная угловая скорость

robot = Roko2(x0, y0, heading0, velocity0, angular_rate0)

dwa_planner = DWAPlanner(config)

# Цикл симуляции
goal_index = 0
goal = waypoints[goal_index]

for _ in range(1000):
    robot_state = robot.get_measurements()
    # Проверяем достижение текущей цели
    distance_to_goal = math.hypot(robot_state.x - goal[0], robot_state.y - goal[1])
    if distance_to_goal < 0.2:
        goal_index += 1
        if goal_index >= len(waypoints):
            print("Робот завершил движение по траектории.")
            break
        goal = waypoints[goal_index]
    # Получаем управление от DWA
    u = dwa_planner.plan(robot_state, goal)
    desired_velocity = u[0]
    desired_angular_rate = u[1]
    # Устанавливаем желаемые параметры движения
    robot.set_motion(desired_velocity, desired_angular_rate)
    # Обновляем состояние робота
    robot.update()

robot.plot_results()
```

```python
# Построение результатов
plt.figure(figsize=(10, 10))
plt.plot(robot._X, robot._Y, 'b', label='Траектория робота')
# Рисуем траекторию цели
waypoints_x = [point[0] for point in waypoints]
waypoints_y = [point[1] for point in waypoints]
plt.plot(waypoints_x, waypoints_y, 'r--', label='Желаемая траектория')
plt.legend()
plt.grid()
plt.xlabel('X, м')
plt.ylabel('Y, м')
plt.title('Движение робота по квадратной траектории с помощью DWA')
plt.axis('equal')
plt.show()
```
