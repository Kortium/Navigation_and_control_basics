```python
import ipywidgets as widg
```

```python
def linef(k, l):
    X = list(range(0,400))
    X = [X[i]/10 for i in range(len(X))]
    Y = [X[i]*k+l for i in range(len(X))]
    matplotlib.pyplot.plot(X,Y)
    matplotlib.pyplot.axis('equal')
    return
widg.interact(linef, k=(0.1,2.0,0.01), l=(1,20))
```
